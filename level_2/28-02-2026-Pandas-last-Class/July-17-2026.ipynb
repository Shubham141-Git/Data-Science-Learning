# Categorical Data in Pandas

### 📌 What is Categorical Data?

Categorical data is textual or labeled data that represents discrete categories.

# Interview Notes

- pd.Categorical() creates a categorical data type.

- Categories are stored once and represented internally by integer codes.

- Reduces memory usage for repeated text values.

- ordered=True enables meaningful comparisons and sorting.

- Commonly used for columns like priority, rating, size, gender, status, etc.

# `pd.Categorical()` in Pandas

`pd.Categorical()` is used to convert data into **categorical data**, which helps save memory and makes operations like sorting, grouping, and comparisons more efficient.

---

# Syntax

```python
pd.Categorical(
    values,
    categories=None,
    ordered=False,
    dtype=None
)
```

---

# Parameters

| Parameter | Required? | Purpose |
|-----------|-----------|---------|
| `values` | ✅ Yes | The data you want to convert into categorical values. |
| `categories` | ❌ No | Specify the allowed categories and their order. |
| `ordered` | ❌ No | Whether the categories have an order (`True` or `False`). |
| `dtype` | ❌ No | Specify a categorical data type (advanced usage). |

---

# 1. `values` (Required)

This is the **data that will become categorical**.

It can be:

- List
- Tuple
- NumPy Array
- Pandas Series

## Example

```python
import pandas as pd

cities = pd.Categorical(
    ["Delhi", "Mumbai", "Delhi", "Pune"]
)

print(cities)
```

### Output

```python
['Delhi', 'Mumbai', 'Delhi', 'Pune']
Categories (3, object): ['Delhi', 'Mumbai', 'Pune']
```

Here,

```python
["Delhi", "Mumbai", "Delhi", "Pune"]
```

is the **`values`** parameter.

---

# 2. `categories`

Suppose your data is:

```python
["Gold", "Silver", "Gold"]
```

You know that the possible categories are:

- Gold
- Silver
- Bronze

You can specify them explicitly.

## Example

```python
grades = pd.Categorical(
    ["Gold", "Silver", "Gold"],
    categories=["Bronze", "Silver", "Gold"]
)

print(grades)
```

### Output

```python
['Gold', 'Silver', 'Gold']
Categories (3, object): ['Bronze', 'Silver', 'Gold']
```

### Notice

Although `"Bronze"` doesn't appear in the data, it is still stored as a valid category.

---

## Why use `categories`?

### Without specifying categories

```python
pd.Categorical(["A", "B", "A"])
```

Pandas automatically discovers:

```text
Categories:
A
B
```

---

### With specifying categories

```python
pd.Categorical(
    ["A", "B", "A"],
    categories=["A", "B", "C", "D"]
)
```

Now Pandas knows that all four values are valid categories, even if some are not present in the data.

---

# 3. `ordered`

By default,

```python
ordered=False
```

This means the categories have **no ranking**.

## Example

```python
grades = pd.Categorical(
    ["High", "Medium", "Low"],
    ordered=False
)
```

Here,

```text
High > Medium
```

has **no meaning**.

---

## Using `ordered=True`

```python
grades = pd.Categorical(
    ["High", "Medium", "Low"],
    categories=["Low", "Medium", "High"],
    ordered=True
)
```

Now Pandas understands:

```text
Low < Medium < High
```

Comparisons now work correctly.

### Example

```python
grades[0] > grades[1]
```

Conceptually,

```text
High > Medium
```

returns

```python
True
```

Sorting also follows the category order.

---

# 4. `dtype`

This is an **advanced parameter**.

Instead of writing:

```python
pd.Categorical(
    values,
    categories=["A", "B"],
    ordered=True
)
```

you can create a reusable categorical data type.

## Example

```python
from pandas.api.types import CategoricalDtype

grade_type = CategoricalDtype(
    categories=["A", "B", "C"],
    ordered=True
)
```

Now reuse it:

```python
pd.Categorical(
    ["A", "B", "A"],
    dtype=grade_type
)
```

This is useful when multiple DataFrames should use the same category definition.

---

# Complete Example

```python
import pandas as pd

grades = pd.Categorical(
    values=["A", "B", "A", "C"],
    categories=["A", "B", "C", "D"],
    ordered=True
)

print(grades)
```

### Output

```python
['A', 'B', 'A', 'C']
Categories (4, object): ['A' < 'B' < 'C' < 'D']
```

# Interview Notes

## `pd.Categorical()` Parameters

| Parameter | Description |
|-----------|-------------|
| `values` | The data to convert into categories. **Required.** |
| `categories` | Defines all possible category values. Optional. |
| `ordered` | Specifies whether the categories have an order (`True`/`False`). Default is `False`. |
| `dtype` | A predefined categorical data type for advanced reuse. Optional. |

---

# Quick Memory Trick

Think of the parameters like this:

```python
pd.Categorical(
    values,        # What is my data?
    categories,    # What are all the allowed values?
    ordered,       # Is there a ranking?
    dtype          # Reuse an existing categorical type (advanced)
)
```

---

# Real-World Usage

In most real-world projects, you'll commonly use these three parameters:

- ✅ `values`
- ✅ `categories`
- ✅ `ordered`

The `dtype` parameter is mainly used when you want multiple DataFrames or Series to share the same categorical definition.

---

# Summary

| Parameter | Purpose | Default |
|-----------|---------|---------|
| `values` | Data to convert into categorical values | **Required** |
| `categories` | Defines all possible categories | `None` |
| `ordered` | Specifies whether categories have an order | `False` |
| `dtype` | Reuse an existing categorical data type | `None` |
