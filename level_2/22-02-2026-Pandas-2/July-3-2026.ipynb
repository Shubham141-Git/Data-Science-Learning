# Series Indexing And Access

````python
import pandas as pd
# 1. Positional Indexing: Accessing elements by their integer position.
data=[1,2,3,4,5 ]
s1=pd.Series(data)
print(s1[0])  # this is more like numpy based indexing # Accessing the first element
print(s1[2])  # Accessing the third element
print(s1.values[4]) # Accessing the fifth element

#iloc gets rows (and/or columns) at integer positions. It is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array.
# iloc: Accessing elements by their integer position using the iloc indexer.
# ILOC STANDS FOR INTEGER LOCATION. It allows you to access elements in a Series or DataFrame based on their integer position, regardless of the index labels.
# iloc is pandas based indexing 
data=[1,2,3,4,5 ]
s1=pd.Series(data)
print(s1.iloc[0])  # Accessing the first element
print(s1.iloc[1:3])  # Slicing operation  to access the second and third elements here 3 is exclusive

# 2. Label-Based Indexing: Accessing elements by their index labels.

# loc gets rows (and/or columns) with particular labels from the index. It is primarily label based, but may also be used with a boolean array.

# Pandas resolves s2[0] in the following order:

# Check if 0 is a label in the index
# → If yes, return by label

# If not, and if the index is not integer-based, then treat 0 as a position
# → Return item at index position 0 (first element)

# FutureWarning: Series.__getitem__ treating keys as positions is deprecated. 
# In a future version, integer keys will always be treated as 
# labels (consistent with DataFrame behavior). 
# To access a value by position, use `ser.iloc[pos]`

s5=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'],name='Scores')
print(s5.loc['b'])  # Accessing the element with label 'b'
print(s5.loc['c':'e'])  # Slicing operation to access elements from label 'c' to 'e' (inclusive)
print(s5['a'])  # Accessing the element with label 'a'
# TO access multiple elements by label, you can use a list of labels:
print(s5.loc[['a', 'c', 'e']])  # Accessing elements with labels 'a', 'c', and 'e'
print(s5[['a','b','c']]) # Accessing elements with labels 'a', 'b', and 'c'

# Only used iloc and loc to access the elements of the series.

````

# 3. What is `iloc`?

`iloc` stands for **Integer Location**.

It is used to select **rows and columns by their integer (position-based) index**, not by their labels.

Think of it like:

- Accessing elements using **index positions**
- Similar to Python list indexing
- Used when you know the position of the data instead of its label

---

## Why `iloc`?

Sometimes, you want to retrieve data based on its position rather than its row or column name.

`iloc` allows you to:

- Select a single row
- Select multiple rows
- Slice rows
- Select a single column
- Select multiple columns
- Select a specific cell
- Select rows and columns together
- Access the last row using negative indexing

---

## Syntax

### For a DataFrame

```python
DataFrame.iloc[row_index, column_index]
```

### For a Series

```python
Series.iloc[index]
```

Where:

- **row_index** → Row position (starts from `0`)
- **column_index** → Column position (starts from `0`)

---

## `iloc` vs `loc`

| Feature | `iloc` | `loc` |
|---------|--------|--------|
| Uses | Integer positions | Row/Column labels |
| Indexing | Starts from 0 | Uses actual index names |
| Similar To | Python list indexing | Dictionary lookup |
| Example | `df.iloc[1, 2]` | `df.loc["B", "Age"]` |

- It always uses **position-based indexing**.
- Row and column positions start from **0**.

- Use `iloc` when you know the **position** of the data, not its label.

# 4. What is `head()` in Pandas?

The `head()` method is used to display the **first few rows** of a **DataFrame** or **Series**.

Think of it like:

- Taking a quick preview of your data
- Looking at the top of an Excel sheet
- Checking whether the data has been loaded correctly

It is commonly used for:

- Inspecting datasets
- Verifying imported data
- Understanding the structure of the DataFrame
- Detecting missing or incorrect values

---

## Why `head()`?

Datasets often contain **thousands or even millions of rows**.

Printing the entire dataset is:

- Slow

Why is columns an attribute and not a method?

A method performs an action, while an attribute stores information.

For example:

df.info()      # Performs an action (prints summary)

These are methods, so they use ().

These are attributes, so they do not use () because they simply provide stored information about the DataFrame.

# 6. What is `loc`?

`loc` stands for **Label Location**.

It is used to select **rows and columns by their labels (names)** instead of their integer positions.

Think of it like:

- Looking up data using **row labels** and **column names**
- Similar to a **dictionary lookup**
- Used when you know the actual labels instead of their positions

---

## Why `loc`?

Sometimes, your DataFrame has meaningful row labels (IDs, names, dates, etc.) instead of numeric indexes.

`loc` allows you to:

- Select a single row
- Select multiple rows
- Select a single column
- Select multiple columns
- Select a specific cell
- Slice rows and columns using labels

---

## Syntax

```python
DataFrame.loc[row_labels, column_labels]
```

Where:

- **row_labels** → Row label(s)
- **column_labels** → Column label(s)

You can provide:

- Only rows
- Rows and columns
- Multiple rows
- Multiple columns
- Label slices

---

## Example DataFrame

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 22],
    "City": ["Pune", "Mumbai", "Delhi"]
}

df = pd.DataFrame(data, index=["A", "B", "C"])

print(df)
```

### Output

```
      Name  Age    City
A    Alice   25    Pune
B      Bob   30  Mumbai
C  Charlie   22   Delhi
```

---

## Understanding Row and Column Labels

### Row Labels

| Label | Name |
|--------|------|
| A | Alice |
| B | Bob |
| C | Charlie |

### Column Labels

| Column Name |
|-------------|
| Name |
| Age |
| City |

Unlike `iloc`, `loc` uses these **labels**, not positions.

30

- Row label = `"B"`

# `loc` vs `iloc`

| Feature | `loc` | `iloc` |
|---------|--------|---------|
| Uses | Labels | Integer Positions |
| Row Selection | Row labels | Row positions |
| Column Selection | Column names | Column positions |
| Slicing | End label included | End position excluded |
| Example | `df.loc["B", "Age"]` | `df.iloc[1, 1]` |

---

## Real-Life Analogy

Imagine a library.

| Method | Analogy |
|---------|----------|
| `iloc` | Find the 2nd book on the shelf |
| `loc` | Find the book by its title |

Similarly,

- `iloc[1]` → Second row
- `loc["B"]` → Row whose label is **B**

---

## Common Syntax

| Syntax | Meaning |
|---------|---------|
| `df.loc[row]` | Select row(s) |
| `df.loc[row, column]` | Select a specific cell |
| `df.loc[:, column]` | All rows, selected column(s) |
| `df.loc[row, :]` | Selected row(s), all columns |
| `df.loc[row_start:row_end, col_start:col_end]` | Slice rows and columns using labels |

---

## Key Points

- `loc` stands for **Label Location**.
- It uses **row labels** and **column names**.
- It supports selecting rows, columns, cells, and slices.
- Label slicing **includes the ending label**.
- It is ideal when your DataFrame has meaningful row labels.

---

## Quick Memory Trick

- **`loc` = Label Location** → Think **labels** (`"A"`, `"Bob"`, `"Age"`).
- **`iloc` = Integer Location** → Think **positions** (`0`, `1`, `2`).

Remember:

- `loc` → Labels
- `iloc` → Positions
