# NumPy Character and String Functions (`np.char`)
This notebook contains examples of common NumPy string operations.

## Import

```
import numpy as np
arr=np.array(['apple','banana','mango'])
```

## upper

```
print(np.char.upper(arr))
```

## lower

```
arr=np.array(['Apple','BANANA','MANGO'])
print(np.char.lower(arr))
```

## capitalize

```
arr=np.array(['apple','banana'])
print(np.char.capitalize(arr))
```

## title

```
arr=np.array(['hello world','data science'])
print(np.char.title(arr))
```

## swapcase

```
arr=np.array(['PyThOn','NuMpY'])
print(np.char.swapcase(arr))
```

## strip/lstrip/rstrip

```
arr=np.array(['  apple  ',' banana '])
print(np.char.strip(arr))
print(np.char.lstrip(arr))
print(np.char.rstrip(arr))
```

## replace

```
arr=np.array(['apple','banana'])
print(np.char.replace(arr,'a','@'))
```

## add

```
a=np.array(['Hello','Good'])
b=np.array([' World',' Morning'])
print(np.char.add(a,b))
```

## multiply

```
arr=np.array(['Hi','Bye'])
print(np.char.multiply(arr,3))
```

## split

```
arr=np.array(['apple mango','banana orange'])
print(np.char.split(arr))
```

## join

```
arr=np.array(['ABC','XYZ'])
print(np.char.join('-',arr))
```

## find

```
arr=np.array(['apple','banana'])
print(np.char.find(arr,'a'))
```

## count

```
arr=np.array(['banana','apple'])
print(np.char.count(arr,'a'))
```

## startswith / endswith

```
arr=np.array(['apple','banana'])
print(np.char.startswith(arr,'a'))
print(np.char.endswith(arr,'e'))
```

## isalpha / isdigit

```
arr=np.array(['apple','abc123'])
print(np.char.isalpha(arr))
arr=np.array(['123','12a'])
print(np.char.isdigit(arr))
```

## islower / isupper

```
arr=np.array(['apple','Apple'])
print(np.char.islower(arr))
arr=np.array(['APPLE','Apple'])
print(np.char.isupper(arr))
```
