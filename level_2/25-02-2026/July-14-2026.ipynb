# Are Subsets Views or Copies in Pandas?

In pandas, whether you get a view (linked to original DataFrame) or a copy (independent) 
depends on how you slice or filter — and it’s not always obvious.

- .iloc always returns a copy (safe, independent subset.)
- .loc may return a view (especially if indexing on labels)
    ```bash
    subset = df.loc[df['Sex'] == 'male']
    This is like the earlier filtering — might be a view.

    so..
    subset = df.loc[df['Sex'] == 'male'].copy()

    ```

| Situation                          | View or Copy? | Should you use `.copy()`? |
| ---------------------------------- | ------------- | ------------------------- |
| `.iloc[]` slicing                  | Mostly Copy   | Optional                  |
| `.loc[]` slicing                   | May be View   | Recommended               |
| Filtering with condition `df[...]` | May be View   | Recommended               |
| You plan to modify the subset      | Use `.copy()` | YES                       |

# Set Index vs Re - index

| Feature       | `set_index()`                               | `reindex()`                                          |
| ------------- | ------------------------------------------- | ---------------------------------------------------- |
| What it does  | **Moves a column** to become the index      | **Changes the row/column labels** of a DataFrame     |
| Main Use Case | Change the row labels using existing column | Match to new labels, add/remove/reorder rows/columns |
| Affects index | Yes – creates a **new index from a column** | Yes – but just **remaps or aligns to new index**     |
| Adds NaNs?    | No – only restructures                      | Yes – can insert NaNs for new labels                 |

````python
import pandas as pd
sales_jan = pd.DataFrame({
    'Sales': [250, 400, 300]
}, index=['Store_A', 'Store_B', 'Store_C'])

sales_jan

print(sales_jan)

sales_feb = pd.DataFrame({
    'Sales': [260, 310]
}, index=['Store_A', 'Store_C'])

sales_feb

print(sales_feb)

# Here both the data frame we can see the index are different jan has Store_A, store_B, Store_c and feb has Store_A and Store_B
````

❌ Problem: Indexes are mismatched!

You cannot safely do operations like subtraction or direct comparison unless indexes are aligned.

✅ 🛠 Solution: Use reindex() to align!

````python
import pandas as pd

# So to align the index of both dataframe we need to know the  which dataframe having full index full index , here we have jan with full indexes present of feb

common_index=sales_jan.index # gives indexes present in jan which is the common index present in 2 dataframe
print(common_index)

# now i want to reindex the dataframe of sales_feb on common index

sales_feb_aligned = sales_feb.reindex(common_index)

sales_feb_aligned # With NaN

'''  
	Sales
Store_A	260.0
Store_B	NaN
Store_C	310.0

'''
# We dont need nan bot zero we use fill_value=o as the parameter in re,index

sales_feb_aligned_with_zero = sales_feb.reindex(common_index, fill_value=0) # here it will directly affects the label and not the column to make the column align we have to explicitly use columns = dataframeName.columns

sales_feb_aligned_with_zero

#Now  jan and feb has the same indexes bow we can persornm the oprewation between both of them

sales_diff = sales_feb_aligned_with_zero['Sales'] - sales_jan['Sales']

sales_diff
````

# Iteration in Pandas

In pandas, iteration refers to looping over rows or columns of a DataFrame or Series. While pandas is optimized for vectorized operations, sometimes we need to loop — for example, for conditional logic, debugging, or generating complex values.

````python
import pandas as pd
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['Delhi', 'Mumbai', 'Chennai']
})

print(df)

for index, row in df.iterrows():
    print(f"{index}: {row['Name']} lives in {row['City']} and is {row['Age']} years old.")
````

````python
for row in df.itertuples():
    print(row)
    print(row.Index)
    print(f"{row.Name} lives in {row.City}, and is {row.Age} years old.", end='\n\n')
````

````python
# Iterating Over Columns

for col in df.columns:
    print(f"Column: {col}")
    print(df[col].values, end='\n\n')
````

# Sorting in Pandas

Sorting helps you organize your data for better readability, analysis, and logical operations — whether by index or actual data values.

### Why Sorting Matters

You might want to:

- See top-performing products (e.g., highest sales)

- Order records by name/date

- Prepare data before merging or visualizing

### Two Main Types of Sorting
1. sort_values() → Sort by data values
2. sort_index() → Sort by index (row or column labels)

# 1. sort_values() → Sort by data values

# syntax

```bash
DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, na_position='last')
```

| Argument      | Purpose                                 |
| ------------- | --------------------------------------- |
| `by`          | Column label(s) to sort by              |
| `axis`        | 0 = rows (default), 1 = columns         |
| `ascending`   | True = low to high, False = high to low |
| `inplace`     | True = modify current object            |
| `na_position` | 'first' or 'last' for NaNs              |


````python
# Example of inplace

class Cal:
    def __init__(self, a):
        self.a = a
    def add(self, b):
        c = self.a + b
        return c
    def inplace_add(self, b):
        self.a = self.a + b
        return self.a
    
c_cal = Cal(10)  # thid method is not updating the value of a by itself , sowe have to perform below operation to update value of a
c_cal.a = c_cal.add(10) # inplace False

# Where below method update the value of a by itself
c_cal.inplace_add(10) # inplace True
````
