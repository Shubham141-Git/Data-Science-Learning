```python
## How to check the condition are indepent or not?
'''
The core rule (memorize this)
‚úÖ Use if when conditions are independent

More than one condition can be True at the same time

‚úÖ Use elif when conditions are mutually exclusive

Only one condition can be True

How to identify which one you need (step-by-step)
üîç Ask yourself ONE question:

Can more than one condition be true at the same time?

YES ‚Üí use multiple if

NO ‚Üí use if / elif / else

That‚Äôs it. Everything flows from this.

Example 1: Pizza toppings (Independent ‚Üí if)
if extra_cheese:
    total += 2

if pepperoni:
    total += 3

if olives:
    total += 1

Why if?

Customer can choose cheese AND pepperoni AND olives

Multiple conditions can be true together

Each adds its own cost

üëâ Independent conditions

Example 2: Traffic light (Exclusive ‚Üí elif)
if signal == "red":
    print("Stop")
elif signal == "yellow":
    print("Slow down")
elif signal == "green":
    print("Go")

Why elif?

A signal can‚Äôt be red and green at the same time

Only one outcome is valid

üëâ Mutually exclusive conditions

A simple mental trick üß†
Replace elif with ‚Äúotherwise if‚Äù
if this happens
otherwise if this happens
otherwise if this happens
otherwise do this


If that sentence makes sense ‚Üí elif
If it sounds wrong ‚Üí use separate if

Common beginner mistake (very important ‚ö†Ô∏è)
‚ùå Wrong
if cheese:
    add 2
elif pepperoni:
    add 3


üëâ This means:

‚ÄúIf cheese is chosen, ignore pepperoni even if it‚Äôs chosen‚Äù

‚úÖ Correct
if cheese:
    add 2

if pepperoni:
    add 3

Decision table (bookmark this)
Situation	Use
Multiple things can apply	if
Only one thing can apply	elif
Additive logic (prices, scores)	if
Choice / selection logic	elif
Feature toggles	if
States / modes	elif
Real-world analogy üåç
üßæ Shopping bill

Item A added ‚Üí price increases

Item B added ‚Üí price increases
üëâ Use if

üö¶ Traffic signal

Red OR Yellow OR Green
üëâ Use elif

Final golden rule üèÜ

If skipping one condition because another is true would be WRONG ‚Äî never use elif.
'''
```
