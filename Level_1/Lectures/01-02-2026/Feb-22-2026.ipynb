## File Operation-2


```python
# Read mode --> 'r'  --> reads the content of file

with open("temp/temp_1.txt", 'r') as f:
    print(f.tell()) # tells the current position of the pointer in file
    data = f.read()
    print(f.tell())
    print(data)
```

```python

# Write mode --> 'w' -->deletes the existing content of file and write new content in it
with open("temp/temp_1_copy.txt", 'w') as f:
    print(f.tell()) # tells the current position of the pointer in file
    data = f.write("Hello World")
    print(f.tell())
    print(data)

#Scenario-2 --> If the file is not present in the directory then it will create a new file and write the content in it. same for w+

with open("temp/Write-NewFile.txt", 'w') as f:
    print(f.tell()) # tells the current position of the pointer in file
    data = f.write("Create new file with write mode")
    print(f.tell())
    print(data)    
```

```python
# Write mode --> 'a' --> appends the new content at the end of file
with open("temp/temp_1_copy.txt", 'a') as f:
    print(f.tell()) # tells the current position of the pointer in file
    f.write(" \n---- Adding the new content at the end of file")
    print(f.tell())

#Scenario-2 -->append will create a new file if the file is not present in the directory   same for a+ 

with open("temp/append-NewFile.txt", 'a') as f:
    print(f.tell()) # tells the current position of the pointer in file
    f.write(" \n---- Adding the new  file")
    print(f.tell())    


  
```

```python
#Read and Write mode --> 'r+' --> reads the content of file and write new content in it without deleting the existing content of file
#for this we have to make use of READ FIRST BECAUSE AFTER READ THE POINTER WILL AT THE END AND FROM THERE WE HAVE TO WRITE THE SECOND CONTENT IN FILE,IF WE DONT USE READ FIRST AND DIRECTLY USE WRITE IT WIPE OUT THE EXISTING CONTENT
with open("temp/temp_1_copy.txt", 'r+') as f:
    print(f.tell()) # tells the current position of the pointer in file
    data = f.read()
    print(data)
    f.write(" \n --Welcome to Python Programming--")
    print(f.tell())

#sCENARIO 2 --> NOT USUBG READ FIRST AND DIRECTLY USING WRITE
#iN THIS CASE IN r+ MODE THE EXISTING CONTENT WILL NOT BE WIPE OUT BUT IT WILL START WRITING FROM THE START i.e FROM THE FIRST POSITION AND IT WILL OVERWRITE THE EXISTING CONTENT (overwrite is the catch)
with open("temp/temp_1_copy.txt", 'r+') as f:
    print(f.tell()) # tells the current position of the pointer in file
    #data = f.read()git stat
    #print(data)
    f.write("*****************")
    print(f.tell())

#Note --> r+ does not create a new file if the file is not present in the directory it will give an error

  
  
```

```python
#What if i want to add the new contenet in middle of the file or at the start of the file 
#for this we have to make use of seek() method to move the pointer to the desired position and then write the new content in file

with open("temp/temp_2", 'w+') as f:
    print(f.tell()) # tells the current position of the pointer in file
    f.write("Hello World")
    print(f.tell())
    f.seek(5) # move the pointer to the start of file
    f.write(" ******")
    print(f.tell())

#Note --> seek() method is used to move the pointer to the desired position in file and then we can write the new content in file from that position, it works in read and write mode but it does not work in append mode because in append mode the pointer is always at the end of file and we cannot move it to any other position in file
#seek does not work in append mode because in append mode the pointer is always at the end of file and we cannot move it to any other position in file    
```

# Exception Handling

```python
# Exception Handling
# Exception handling is a mechanism in programming used to handle runtime errors, so that the program does not crash and can continue executing normally.
#There might be some flaky code which might cause the  creash i.e system will stop the execution at that point  and it will not execute the remaining code, to handle such type of situation we can use exception handling mechanism in python.
#In python we have try and except block to handle the exceptions, we can write the code

#example 

userInput = input("Enter a number: ")  
def addTwoNumbers(num1,num2):
    return num1 + num2

'''
addTwoNumbers(10, "hello") # this will give an error because we cannot add a number and a string together, it will give a TypeError: unsupported operand type(s) for +: 'int' and 'str' and the next line of code will not be executed 
print("Hello World") # this line will not be executed because the previous line of code gives an error and the program will stop executing at that point
'''

#To handle this type of situation we can use try and except block to handle the exception and the program will continue executing normally without crashing
try:
    addTwoNumbers(10, "hello") # this will give an error because we cannot add a number and a string together, it will give a TypeError: unsupported operand type(s) for +: 'int' and 'str' and the next line of code will not be executed  
except TypeError as e:
    print("Error:", e)
print("Hello World") # this line will be executed because the exception is handled
```
