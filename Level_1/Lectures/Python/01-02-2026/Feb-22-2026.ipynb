## File Operation-2


```python
# Read mode --> 'r'  --> reads the content of file

with open("temp/temp_1.txt", 'r') as f:
    print(f.tell()) # tells the current position of the pointer in file
    data = f.read()
    print(f.tell())
    print(data)
```

```python

# Write mode --> 'w' -->deletes the existing content of file and write new content in it
with open("temp/temp_1_copy.txt", 'w') as f:
    print(f.tell()) # tells the current position of the pointer in file
    data = f.write("Hello World")
    print(f.tell())
    print(data)

#Scenario-2 --> If the file is not present in the directory then it will create a new file and write the content in it. same for w+

with open("temp/Write-NewFile.txt", 'w') as f:
    print(f.tell()) # tells the current position of the pointer in file
    data = f.write("Create new file with write mode")
    print(f.tell())
    print(data)    
```

```python
# Write mode --> 'a' --> appends the new content at the end of file
with open("temp/temp_1_copy.txt", 'a') as f:
    print(f.tell()) # tells the current position of the pointer in file
    f.write(" \n---- Adding the new content at the end of file")
    print(f.tell())

#Scenario-2 -->append will create a new file if the file is not present in the directory   same for a+ 

with open("temp/append-NewFile.txt", 'a') as f:
    print(f.tell()) # tells the current position of the pointer in file
    f.write(" \n---- Adding the new  file")
    print(f.tell())    


  
```

```python
#Read and Write mode --> 'r+' --> reads the content of file and write new content in it without deleting the existing content of file
#for this we have to make use of READ FIRST BECAUSE AFTER READ THE POINTER WILL AT THE END AND FROM THERE WE HAVE TO WRITE THE SECOND CONTENT IN FILE,IF WE DONT USE READ FIRST AND DIRECTLY USE WRITE IT WIPE OUT THE EXISTING CONTENT
with open("temp/temp_1_copy.txt", 'r+') as f:
    print(f.tell()) # tells the current position of the pointer in file
    data = f.read()
    print(data)
    f.write(" \n --Welcome to Python Programming--")
    print(f.tell())

#sCENARIO 2 --> NOT USUBG READ FIRST AND DIRECTLY USING WRITE
#iN THIS CASE IN r+ MODE THE EXISTING CONTENT WILL NOT BE WIPE OUT BUT IT WILL START WRITING FROM THE START i.e FROM THE FIRST POSITION AND IT WILL OVERWRITE THE EXISTING CONTENT (overwrite is the catch)
with open("temp/temp_1_copy.txt", 'r+') as f:
    print(f.tell()) # tells the current position of the pointer in file
    #data = f.read()git stat
    #print(data)
    f.write("*****************")
    print(f.tell())

#Note --> r+ does not create a new file if the file is not present in the directory it will give an error

  
  
```

```python
#What if i want to add the new contenet in middle of the file or at the start of the file 
#for this we have to make use of seek() method to move the pointer to the desired position and then write the new content in file

with open("temp/temp_2", 'w+') as f:
    print(f.tell()) # tells the current position of the pointer in file
    f.write("Hello World")
    print(f.tell())
    f.seek(5) # move the pointer to the start of file
    f.write(" ******")
    print(f.tell())

#Note --> seek() method is used to move the pointer to the desired position in file and then we can write the new content in file from that position, it works in read and write mode but it does not work in append mode because in append mode the pointer is always at the end of file and we cannot move it to any other position in file
#seek does not work in append mode because in append mode the pointer is always at the end of file and we cannot move it to any other position in file    
```

# Exception Handling

```python
# Exception Handling
# Exception handling is a mechanism in programming used to handle runtime errors, so that the program does not crash and can continue executing normally.

#Example

userInput = input("Enter a number: ")
def addTwoNumbers(num1,num2):
    return num1 + num2

addTwoNumbers(10, "hello") # this will give an error because we cannot add a number and a string together, it will give a TypeError: unsupported operand type(s) for +: 'int' and 'str' and the next line of code will not be executed
print("Hello World") # this line will not be executed because the previous line of code gives an error and the program will stop executing at that point

#To habdle we can make use of try except block to handle the exception and the program will continue executing normally without crashing
try:    
    addTwoNumbers(10, "hello") # this will give an error because we cannot add a number and a string together, it will give a TypeError: unsupported operand type(s) for +: 'int' and 'str' and the next line of code will not be executed
except Exception as e:
    print("An error occurred: ", e)         
print("Hello World") # this line will be executed because the exception is handled and the program will continue executing normally without crashing

#Note --> we can also use multiple except block to handle different types of exceptions separately and we can also use else block to execute the code when there is no exception and finally block to execute the code that will be executed regardless of whether there is an exception or not.



'''We Write the risky code inside the try block and if there is any exception occurs then it will be caught by the except block 
and we can handle the exception in the except block and the program will continue executing normally without crashing. '''


'''What except Exception as e: Means
1️⃣ Exception

Exception is a base class for most errors in Python.

It catches almost all standard runtime errors.

Examples:

ValueError

TypeError

ZeroDivisionError

FileNotFoundError

So writing:

except Exception:

means 👉 “Catch almost any error that happens.”

2️⃣ as e

This stores the error object inside a variable (e).

So:

except Exception as e:

means:

“Catch the error and store its details in variable e.”

🔹 Example
try:
    x = 10 / 0
except Exception as e:
    print("Error occurred:", e)
Output:
Error occurred: division by zero

Here:

The error is ZeroDivisionError

It gets caught by Exception

e contains the message "division by zero"

🔹 Why Use as e?

It helps you:

Print error messages

Log errors

Debug problems

Show custom error messages

Example:

except Exception as e:
    print(f"Something went wrong: {e}")

##############################################

'''
```

# Try except else

```python
# ELSE BLOCK WILL ONLY EXECUTE WHEN THE TRY BLOCK IS EXECUTED WITHOUT ANY EXCEPTION AND IT WILL NOT EXECUTE IF THERE IS ANY EXCEPTION OCCURS IN TRY BLOCK
#ELSE BLOCK IS OPTIONAL

try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    
    result = num1 / num2

except ZeroDivisionError:
    print("❌ You cannot divide by zero!")

except ValueError:
    print("❌ Please enter valid integers!")

else:
    print("✅ Division successful!")
    print("Result:", result)
```

# TRY EXCEPT FINALLY

```python
# FINALLY BLOCK WILL RUN NO MATTER WHAT WHETHER THERE IS AN EXCEPTION OR NOT, IT WILL ALWAYS EXECUTE, IT IS USED TO RELEASE THE RESOURCES OR TO CLOSE THE FILES OR TO CLOSE THE DATABASE CONNECTIONS ETC. IT IS ALSO OPTIONAL   
#

try:
    file = open("data.txt", "r")
    content = file.read()

except FileNotFoundError:
    print("❌ File not found!")

except Exception as e:
    print("❌ Some other error occurred:", e)

else:
    print("✅ File read successfully!")
    print(content)

finally:
    print("📌 This block always runs.")
    try:
        file.close()
    except:
        pass
```

```python
##raise keyword is used to raise an exception manually in the code, it is used when we want to raise an exception based on some condition or when we want to create our own custom exception.   

''' 
What is raise in Python?

The raise keyword is used to manually trigger (throw) an exception in your code.

Normally, Python raises errors automatically (like ZeroDivisionError), but with raise, you can create your own errors or re-throw existing ones.

🔹 Basic Syntax
raise ExceptionType("error message")
🔹 Example 1: Raising a Custom Error
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")

👉 Output:

ValueError: Age cannot be negative
🔹 Example 2: Using raise inside try-except
try:
    num = int(input("Enter a number: "))
    if num < 0:
        raise ValueError("Negative numbers not allowed")
except ValueError as e:
    print("Error occurred:", e)

👉 Here:

You manually raise ValueError
It is caught in except
🔹 Example 3: Re-raising an Exception
try:
    x = int("abc")
except ValueError:
    print("Handling error...")
    raise   # re-throws the same exception

👉 Output:

Handling error...
ValueError: invalid literal for int()
🔹 Example 4: Custom Exception Class
class MyError(Exception):
    pass

raise MyError("This is my custom error")
🔹 Why use raise?

Use raise when:

You want to validate input
You want to stop execution with a meaningful error
You want to enforce rules in your program
You want to re-throw exceptions after partial handling
🔹 Real-life Analogy

Think of raise like:

🚨 Pressing an alarm button manually when you detect a problem

🔹 Key Points
raise = manually trigger error
Can raise built-in or custom exceptions
Can re-throw current exception using just raise
Often used with try-except
'''
```
