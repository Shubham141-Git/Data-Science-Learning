# Type casting
- If we have int and we have to make it float we use type casting

- Type casting is mainly used to change the data type of an varaible

```python
a=1
print(type(a))
```

```python
float(a)
```

```python
print(a) # her it will print only 1 and not 1.0 because we did not updated the value of a
```

```python
# now we update tehvalue of a
a=float(a)
print(a) # now it will print 1.0
```
