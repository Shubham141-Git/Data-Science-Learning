```python
# EVERRYRHING IN PYTHON IS AN OBJECT

def add(x, y):
    print(x + y)
  #here python will create a function object in memory and assign it to the name 'add'  

add(2, 3)  # this will call the function object that 'add' refers to and return 5  

#here add is an address of the memoery location where the function object is stored. When we call add(2, 3), Python looks up the name 'add', finds the function object, and executes it with the provided arguments.
a=id(add)  # this will return the memory address of the function object that 'add' refers to
print
# now we can assign the function object to another name
x = add  # now x also refers to the same function object as add i.e function object is stored at the same memory location and both add and x point to it
#here function is store to the new varaible x, and both add and x refer to the   same  memory location. So when we call x(2, 3), it will execute the same function as add(2, 3) and return 5.
b=id(x)  # this will return the same memory address as id(add) since x and add refer to the same function object
print(b)
print(a == b)  # this will print True since both a and b are the same memory address
x(1,3)
```

# Returning a function from function--> decorator

```python
def add_3():
    print("Addition Function")
```

```python
def my_func(func): # my_func is a decorator function that takes a function as an argument and returns a new function that wraps the original function with some additional functionality.

    def wrap():
        print("Start")
        func()
        print("End")
    return wrap # memory location function object of wrap
```

```python
# add_3 inside the argument below is having different address then the add we are catching
add_3 = my_func(add_3) #here we are catching the the function into new variable which is returned by the my_func and now add_3 is referring to the wrap function object that is returned by my_func. So when we call add_3(), it will execute the code inside wrap, which includes calling the original add_3 function and printing "Start" and "End".
```

```python
add_3()  # here we are calling the wrap function that is returned by my_func, and it will execute the code inside wrap, which includes calling the original add function and printing "Start" and "End".    
```

```python
import time

def my_decorator(func):

    def wrapper():
        t1 = time.perf_counter()
        func()
        t2 = time.perf_counter()
        total_time_takes = t2-t1
        print(f"Function took {total_time_takes:.2f} seconds")
    return wrapper
```

```python
def loop_and_multiply():
    total = 1
    for i in range(1,100):
        total += (total*i)
        time.sleep(0.1)
    print(total)
```

```python
loop_and_multiply = my_decorator(loop_and_multiply) # decorator with manual way
```

```python
loop_and_multiply()
```

```python
# decorator with syntactic sugar

# instead of writing

loop_and_multiply = my_decorator(loop_and_multiply) 

# Python gives shorthand way
@my_decorator # this is already define above , we are just using @functionName--> which calls the function inside it 
def loop_and_multiply():
    total = 1
    for i in range(1,100):
        total += (total*i)
        time.sleep(0.1)
    print(total)

loop_and_multiply()    
```

```python
# Step-1 : Create a decorator

def my_decorator(func):

    def wrapper():
        t1 = time.perf_counter()
        func()
        t2 = time.perf_counter()
        total_time_takes = t2-t1
        print(f"Function took {total_time_takes:.2f} seconds")
    return wrapper

# Step-2 : Create base function

def loop_and_multiply():
    total = 1
    for i in range(1,100):
        total += (total*i)
        time.sleep(0.1)
    print(total)

# Step-3 : Using short hand way, add @my_decorator (@decorator_name) on top of your base function
@my_decorator # create once and use it with multiple function--> which will return the new function with extra functionality
def loop_and_multiply():
    total = 1
    for i in range(1,100):
        total += (total*i)
        time.sleep(0.1)
    print(total)

loop_and_multiply()    
```

```python
@my_decorator
def my_function():
    print("This is my function")    
my_function()    
```

```python
#usecase of decorator --> to handle exception in a function

def decorator_exep(func):
    def wrapper():
        try:
            func()
        except Exception as e:
            print(e)
    return wrapper
```

```python
@decorator_exep
def add():
    a = input("Enter a number : ")
    print(a + 100)
```

```python
#Now what if the arguments are more than 2 , since argument s can be of any number, we can use *args and **kwargs to handle variable number of arguments in the wrapper function.

def decorator_exep_v2(func):
    def wrapper(*args ):
        try:
            func(*args)
        except Exception as e:
            print(e)
    return wrapper

#Now decorator with *args can take any number of positional arguments and pass them to the original function func. This way, we can handle functions with varying numbers of arguments without modifying the decorator for each specific case.

def addition(a,b,c):
 d=a+b+c
 print(d)

addition=decorator_exep_v2(addition) 
addition(2,3,4)

#There is a catch with thos that this does not habndle the keyword arguments, so we can modify the wrapper function to handle both *args and **kwargs
```

```python
@decorator_exep_v2
def add(*args, **kwargs):
    sum = 0
    for i in args:
        sum+=i
    for key, value in kwargs.items():
        sum+=value
    # print(sum)
    return sum

add(10, 20, a=30, b=40) # this will return 100 because it will add all the positional arguments and keyword arguments together and return the sum.
''' 

Let's understand this line step-by-step:

for key, value in kwargs.items():

inside this function:

def add(*args, **kwargs):

    sum = 0

    for i in args:
        sum += i

    for key, value in kwargs.items():
        sum += value

    return sum
First Understand kwargs

**kwargs stores all keyword arguments in a dictionary.

Example call:

add(10, 20, a=30, b=40)

Then Python stores:

args = (10, 20)

kwargs = {
    'a': 30,
    'b': 40
}
What is .items() ?

Dictionary method:

dictionary.items()

returns:

(key, value)

pairs.

Example
data = {
    'a': 30,
    'b': 40
}

print(data.items())

Output:

dict_items([('a', 30), ('b', 40)])
So This Loop Means
for key, value in kwargs.items():

means:

👉 take each key-value pair from dictionary one by one.

Dry Run of Your Code

Function call:

add(10, 20, a=30, b=40)
Step 1 → args
args = (10, 20)

Loop:

for i in args:

Iteration:

i	sum
10	10
20	30
Step 2 → kwargs
kwargs = {
    'a': 30,
    'b': 40
}

Now loop:

for key, value in kwargs.items():
Iteration 1
key = 'a'
value = 30

Then:

sum += value
sum = 30 + 30 = 60
Iteration 2
key = 'b'
value = 40

Then:

sum += value
sum = 60 + 40 = 100
Final Return
return 100
Visualization
kwargs = {
    'a': 30,
    'b': 40
}

.items() converts into:

('a', 30)
('b', 40)

Then loop extracts:

key   value
----  -----
'a'    30
'b'    40
Why Use key ?

Sometimes we need parameter names too.

Example:

for key, value in kwargs.items():

    print(key)
    print(value)

Output:

a
30
b
40
If You Only Need Values

You can also write:

for value in kwargs.values():
    sum += value

Because only values are needed for addition.

Simple Definition
kwargs.items()

returns all dictionary key-value pairs one by one.

'''
```

# Static
- STATIC BELONG TO CLASS NOT AN OBJECT
- STATIC MEANS ITS SHARED ACCROSS ALL OBJECT
- IT IS ONLY CREATED ONCE
- NOT TIED TO ANY INSTANCE

```python
class User:
    # static variable
    course_name = "UDSB2"

    def __init__(self, batch_no):
        self.batch_number = batch_no
```

```python
obj_user_1 = User(1)
obj_user_2 = User(2)

#now calling batch_number for both objects
print(obj_user_1.batch_number) # this will print 1 because we are creating an object of the User class and passing
print(obj_user_2.batch_number) # this will print 2 because we are creating another object of the User class and passing different batch number. Each object has its own instance variable batch_number, so obj_user_1.batch_number will print 1 and obj_user_2.batch_number will print 2.
#So here we can say the batch_number does change with each object because it is an instance variable, and each object has its own copy of the instance variable. Therefore, when we create obj_user_1 with batch number 1, it has its own batch_number variable set to 1, and when we create obj_user_2 with batch number 2, it has its own batch_number variable set to 2. Hence, the batch_number changes with each object.
#But if we call course_name for both objects, it will print the same value because course_name is a static variable and it is shared among all instances of the class. So obj_user_1.course_name and obj_user_2.course_name will both print "UDSB2".
#if any of the object changes the value of course_name, it will change for all objects because it is a static variable and shared among all instances of the class. So if we do obj_user_1.course_name = "New Course", then obj_user_2.course_name will also reflect this change and print "New Course" when accessed. This is because both obj_user_1 and obj_user_2 refer to the same class variable course_name, and any change to it will affect all instances that access it.
print(obj_user_1.course_name) # this will print "UDSB2" because course_name is a static variable and it is shared among all instances of the class. So obj_user_1.course_name and obj_user_2.course_name will both print "UDSB2".
print(obj_user_2.course_name) # this will also print "UDSB2" for
# Note
# 1. Standard method to access static variable is through class name like User.course_name, but it can also be accessed through object reference like obj_user_1.course_name or obj_user_2.course_name. However, it is recommended to access static variables through the class name to make it clear that it is a class variable and not an instance variable.

#object variable can be created inside the constructor or outside the constructor, but it is recommended to create object variables inside the constructor to maintain consistency and clarity in the code. Creating object variables inside the constructor allows us to initialize them when an object is created, and it makes it clear that these variables are specific to each instance of the class. On the other hand, creating object variables outside the constructor can lead to confusion and potential issues with variable scope and initialization. Therefore, it is best practice to define object variables within the constructor of a class.
# example of creating object variable inside the constructor
class User:
    # static variable
    course_name = "UDSB2"

    def __init__(self, batch_no):
        self.batch_number = batch_no # this is an instance variable because it is specific to each instance of the class and can have different values for different objects. Each object will have its own copy of the batch_number variable, and it can be initialized with different values when creating objects of the User class.

#outside the constructor
obj_user_1 = User(1)    
obj_user_1.name = "Shubham" # this is also an instance variable because it is specific to the obj_user_1 instance and can have a different value for another instance of the User class. Each object can have its own set of instance variables, and they can be created and assigned values outside the constructor as well. However, it is generally recommended to define instance variables within the constructor for better organization and clarity in the code.        
```

# static methods : Used for logical group
- no object required
- cleaner name calling

```python
class MathUtils:

    @staticmethod # decorator used to make the method static, which means it can be called on the class itself without needing an instance of the class. Static methods do not have access to instance variables or the self parameter, and they are typically used for utility functions that do not require any information about the instance of the class.
    def add(a,b):
        return a + b
    
    @staticmethod
    def subtract_a_from_b(a,b):
        return a - b

    @staticmethod
    def subtract_b_from_a(a,b):
        return b - a
```
