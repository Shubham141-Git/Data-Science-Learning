```python
import time

t1= time.time() #--> it tels me current excat time, will provide unit of seconds from january 1, 1970(unix time)
print(t1)
```

```python
#perf_counter()--> it will give me the time in seconds but it is more precise than time.time() and it is used to measure the performance of the code

p1= time.perf_counter()
p1
```

```python

def add(a,b):
    print(a+b)
```

```python
#Difference between time.time() and time.perf_counter() is that time.time() gives you the current time in seconds since the epoch (January 1, 1970), while time.perf_counter() provides a higher resolution timer that is more accurate for measuring short durations of time, making it ideal for performance testing and benchmarking.

#example


t1_f = time.time()
add(1,2)
t2_f = time.time()

print(t2_f - t1_f) # give resulr as 0.0 seconds

t1_f_pf = time.perf_counter() # higher precision
add(1,2)
t2_f_pf = time.perf_counter()

print(t2_f_pf - t1_f_pf) # gives result as 0.003111 which is more precise

#Here perf_counter() is more accurate than time.time() for measuring the execution time of the add function, especially when the execution time is very short.  
```

```python
# if i want to wait for 5 seconds then i can use time.sleep(5) which will pause the execution of the code for 5 seconds
print("Hello")
time.sleep(5)
print("World")
```

```python
t1_f_pf = time.perf_counter() # higher precision
time.sleep(5)
t2_f_pf = time.perf_counter()
print(t2_f_pf - t1_f_pf) # gives result as 5.000123 which is more precise
```

```python
#fstring-->  formatted string literals, also known as f-strings, are a way to embed expressions inside string literals, using a minimal syntax. They were introduced in Python 3.6 and provide a more concise and readable way to format strings compared to older methods like str.format() or the % operator.

# usally we write we write
var_1 = "world"
var_2 = "Hello " + var_1
print(var_2)

var_2new = f"Hello {var_1}" # dynamic value is enclosed in {} and it start with f before the string
print(var_2new)

#if we want to show n number of decimal places then we can use :.nf where n is the number of decimal places we want to show
pi = 3.141592653589793
print(f"Pi is approximately {pi:.2f}") # Output: Pi is approximately

#.n% is used to format a number as a percentage with n decimal places. For example, if you want to format a number as a percentage with 2 decimal places, you can use the following syntax:
number = 0.12345    
print(f"The percentage is {number:.2%}") # Output: The percentage is 12.35%

# round() function is a built-in function in Python that rounds a number to a specified number of decimal places. The syntax for the round() function is as follows:
#round(number, ndigits) 
#number: The number you want to round.
#ndigits: The number of decimal places to round to (optional). If not provided,
# the function will round to the nearest integer.
rounded_number = round(pi, 3)
print(rounded_number) # Output: 3.142
```

```python
# polymorphism--> it is the ability of an object to take on many forms. In Python, polymorphism allows us to define methods in a child class that have the same name as methods in the parent class. This means that we can use the same method name to perform different tasks based on the context of the object that is calling the method.
```

```python
class TV:
    def start(self):
        print("TV started")


class Microwave:
    def start(self):
        print("Microwave started")


class Fan:
    def start(self):
        print("Fan started")


class AirFrier:
    def start_AirFrier(self): #--> this is differet name so to avoid it we can make use of abstract class and inheritance
        print("AirFrier started")


 #creating the list of objects
devices = [TV(), Microwave(), Fan(), AirFrier()]
for device in devices:
    device.start()       
```

```python
from abc import ABC, abstractmethod
```

```python
class ElectronicContract(ABC):

    @abstractmethod
    def start(self):
        pass
```

```python

class TV(ElectronicContract):
    def start(self):
        print("TV")

class Microwave(ElectronicContract):
    def start(self):
        print("MV")

class Fan(ElectronicContract):
    def start(self):
        print("Fan")

class AirFrier(ElectronicContract):
    def start_airfried(self):
        print("AirFrier")


devices = [TV(), Microwave(), Fan(), AirFrier()]
```

```python
#Fixing the actual problem


class TV(ElectronicContract):
    def start(self):
        print("TV")

class Microwave(ElectronicContract):
    def start(self):
        print("MV")

class Fan(ElectronicContract):
    def start(self):
        print("Fan")

class AirFrier(ElectronicContract):
    def start(self): #--> now it is same name as other class so it will not create any problem
        print("AirFrier")


devices = [TV(), Microwave(), Fan(), AirFrier()]
for device in devices:
    device.start()
```

# MODULES 
- Python file containing code that can be imported like we used "import OS" but here we will be creating our own class/function which can be imported

- A module in Python is simply a file that contains:

functions,
variables,
classes,
reusable code.

- Instead of writing everything in one file, we divide code into modules.

- Module is always created in .py file
- we create module i.e py files t0 seprate complex, resuable code from the main python file


```python
'''   
Real Life Example

Think of modules like apps in your phone:

App	Purpose
Calculator	calculations
Camera	photos
WhatsApp	messaging

Similarly in Python:

Module	Purpose
math	maths functions
random	random numbers
os	operating system work
datetime	date & time
Example of Using Module
Using math module
import math

print(math.sqrt(25))
print(math.pi)

Output:

5.0
3.141592653589793
Why Modules Are Used

Modules help in:

code reuse,
clean code,
organization,
avoiding duplicate code.
Types of Modules
1. Built-in Modules

Already available in Python.

Examples:

math
os
random
datetime
2. User-defined Modules

Modules created by us.

Creating Your Own Module
Step 1 → Create file

calculator.py

def add(a, b):
    return a + b

def sub(a, b):
    return a - b
Step 2 → Use it in another file

main.py

import calculator

print(calculator.add(10, 5))

Output:

15
Different Ways to Import
1. import module
import math

print(math.sqrt(16))
2. from module import function
from math import sqrt

print(sqrt(16))
3. import with alias
import math as m

print(m.sqrt(25))
Important Built-in Modules
Module	Use
math	mathematical functions
random	random values
os	file/folder handling
sys	Python system functions
datetime	date & time
time	delays/time
json	JSON handling
Example of random
import random

print(random.randint(1, 10))

Random number between 1 and 10.

Example of os
import os

print(os.getcwd())

Gets current folder path.

Module vs Package
Module	Package
Single .py file	folder containing modules
Example: math.py	Example: numpy
Simple Definition

A module is a Python file containing reusable code like functions, classes, and variables.

Memory Trick
Module = Reusable Python File
'''
```
