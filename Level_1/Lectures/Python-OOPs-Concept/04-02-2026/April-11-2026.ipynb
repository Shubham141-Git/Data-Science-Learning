```python
#Creating the class in python using class keyword and the name of the class should be in camel case and it should start with capital letter
class OurBankAccount:
    def __init__(self, name, balance): # This is constructor method which is called when we create an object of the class # constructor helps for memory management and it is used to initialize the variables of the class and it is called when we create an object of the class
        self.name = name 
        self.balance = balance

# __init__(self, name, balance) here name , balance are local to init method but self.name and self.balance are global to the class and can be accessed by any method of the class and self.name and self.balance are instance variables of the class and they are used to store the data of the object and they are used to access the data of the object and they are used to modify the data of the object and they are used to differentiate between the local variable and instance variable of the class

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient Balance")
        else :
            self.balance -= amount
        print(self.name,"withdrew :",amount,". New balance:", self.balance)

        # In Java wwe makeuse of the this keyword to refer to the current object(variables declared inside the class) and in python we use self keyword to refer to the current object(varaible)   
        # In java constructure name is same as method name , but in python constructor name is always __init__ and it is a special method which is called when we create an object of the class  and it is not compulsory to use __init__ method in python but it is a good practice to use it for memory management 
        # In python we use self keyword to refer to the current object and in java we use this keyword to refer to the current object
        # In python constructor the first parameter is always self(can be any name but convention is to use self) and in java constructor there is no need to pass the this keyword as parameter because it is automatically passed by the compiler
        # Self is used to access the variables and methods of the class and it is a reference to the current object of the class and it is used to access the variables and methods of the class and it is used to differentiate between the local variable and instance variable of the class
        # Self point out to the object , which uses the parameter paased into it
#user2 = OurBankAccount("Ganesh", 100000000000) --> here self will point to user2 object and will access the name and balance variable of the user2 object and will perform the withdraw operation on the user2 object


# Note--> Self tells class about it own body/connection--> self compulsorily should be the frist parammeter of every method created inside the class be cause it tells python that the methods are linked to the class and
# Self is the instance of class--> means that self is the object of the class and it is used to access the variables and methods of the class and it is used to differentiate between the local variable and instance variable of the class
#It meNS SELF HAS THE SPECIFIC MEMORY ASSOCIATED WHICH CONTAINS THE CLASS METHOD N VARIABLES AND 
       #self connects methods to the object, allowing access to the object's instance variables.
        '''
the same function is written in java is as follows:

        class OurBankAccount {
    String name;
    double balance;

    OurBankAccount(String name, double balance) {
        this.name = name;
        this.balance = balance;
    }

    void withdraw(double amount) {
        if (amount > this.balance) {
            System.out.println("Insufficient Balance");
        } else {
            this.balance -= amount;
        }
        System.out.println(this.name + " withdrew: " + amount + ". New balance: " + this.balance);
    }
} 
        '''


 # creating objects of the class--> means we r creating the memory location in RAM for the class with its all the method and variable define into class
user1 = OurBankAccount("Shubham", 100000000000)
user2 = OurBankAccount("Ganesh", 100000000000) 

#here now user1 and the user2 are the objects of the class OurBankAccount and we can call the withdraw method for each object   
user1.withdraw(1000)



''' 
Here the varaible (user1) is now bind with function (withdraw) and the data (name and balance) is also bind with the function (withdraw) in the form of an object, and this is called encapsulation in OOPs concept.
now data and function are bind together in the form of an object and this is called encapsulation in OOPs concept.
'''

''' 
🔹 Why withdraw uses self
def withdraw(self, amount):

👉 self represents the object that is calling the method

So if you do:

user1.withdraw(200)

👉 Then inside the method:

self = user1
🔹 How it actually works (step-by-step)
You write:
user1.withdraw(200)
Python internally does:
OurBankAccount.withdraw(user1, 200)

👉 So:

self → user1
amount → 200
🔹 Now your method with real meaning
def withdraw(self, amount):
    if amount > self.balance:
        print("Insufficient Balance")
    else:
        self.balance -= amount
    print(self.name, "withdrew:", amount, ". New balance:", self.balance)
Replace self with user1 mentally:
if amount > user1.balance:
    print("Insufficient Balance")
else:
    user1.balance -= amount

print(user1.name, "withdrew:", amount, ". New balance:", user1.balance)

👉 Now it becomes very clear!

🔹 Why self is necessary

Without self, Python won’t know which object’s data to use

❌ Wrong:
def withdraw(amount):
    balance -= amount   # ❌ which balance?

👉 Problem:

No connection to object
Breaks OOP concept
🔹 OOP concept behind this
🔸 Encapsulation
Data (name, balance) is stored inside object
Methods must access it via self
🔸 Object-specific behavior

Each object behaves independently:

user1 = OurBankAccount("Shubham", 1000)
user2 = OurBankAccount("Rahul", 2000)

user1.withdraw(200)

👉 Only user1 changes because:

self = user1
🔹 Memory view
user1 ──► Object1 (name: Shubham, balance: 1000)
user2 ──► Object2 (name: Rahul, balance: 2000)

Call:

user1.withdraw(200)

👉 Inside method:

self ──► Object1

✔ So only Object1 is modified

🔹 Important rule

👉 Any time you want to:

access object data → use self.variable
modify object data → use self.variable
🔥 One-line understanding

self is the reference to the current object, and it allows methods to access and modify that object's data.

✅ Final takeaway
self = current object (like this in Java)
Required to access instance variables
Ensures correct object is modified'''

''' 
🔹 Why we use self in constructor (__init__)
def __init__(self, name, balance):
    self.name = name
    self.balance = balance

👉 Here self means: “this object being created”

When you do:

user1 = OurBankAccount("Shubham", 1000)

Python internally does:

OurBankAccount.__init__(user1, "Shubham", 1000)

So:

self = user1
self.name → stored inside that object
self.balance → stored inside that object

👉 Without self, the values won’t go into the object.

❌ If you don’t use self
def __init__(self, name, balance):
    name = name
    balance = balance

👉 These become local variables only
👉 Object will have no data stored

🔹 Why self is used in withdraw
def withdraw(self, amount):
    self.balance -= amount

When you call:

user1.withdraw(200)

Python does:

OurBankAccount.withdraw(user1, 200)

👉 So:

self = user1
You are modifying that specific object’s balance
❌ Without self
def withdraw(amount):
    balance -= amount   # ❌ which balance?

👉 Problem:

Python doesn’t know which object’s balance to change
Breaks OOP principle
🔹 Is self compulsory?
✔ Technically:
Python does NOT force the name self
But the first parameter is compulsory

You could write:

def withdraw(this, amount):
    this.balance -= amount

👉 This works, but ❌ bad practice

✔ Practically (important):
self is standard convention
Everyone uses it
Interviewers expect it
🔹 OOP reason (most important)

👉 self is required because of:

🔸 1. Encapsulation
Data belongs to object
Access via self
🔸 2. Object identity
Tells Python which object is calling the method
🔸 3. Instance variables
Created and accessed using self
🔥 Simple analogy

Think:

self = "this particular account"

So:

self.balance -= amount

means:
👉 “Reduce balance of THIS account”

✅ Final takeaway
self connects method → object
Required to store and modify object data
First parameter is compulsory in instance methods
Name self is convention but should always be used
🔥 One-line exam answer:

self is used to refer to the current object and is required to access and modify instance variables inside constructors and methods.

'''
```
