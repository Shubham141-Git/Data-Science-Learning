```python
# __init()__  --> constructor in python, which is a special method that is automatically called when an object of a class is created. It is used to initialize the attributes of the class and set up any necessary state for the object. The __init__ method takes at least one parameter, self, which refers to the instance of the class being created. Additional parameters can be added to __init__ to allow for more flexible object creation.
# __str()__  --> This is a special method in Python that is used to define the string representation of an object. When you call the str() function on an object, or when you print an object, Python will look for the __str__ method to determine how to represent the object as a string. The __str__ method should return a human-readable string that describes the object in a meaningful way. If the __str__ method is not defined, Python will use the default string representation of the object, which is usually not very informative.  
# __repr()__  --> This is a special method in Python that is used to define the official string representation of an object. It is called by the repr() built-in function and should return a string that ideally could be used to recreate the object. If the __repr__ method is not defined, Python will use the default representation, which usually includes the class name and memory address.
```

```python
# str and repr are both special methods in Python that are used to define the string representation of an object, but they serve different purposes. The __str__ method is intended to provide a human-readable representation of the object, while the __repr__ method is intended to provide an unambiguous representation of the object that can be used for debugging and development purposes. If both methods are defined, str() will use __str__, while repr() will use __repr__. If only __repr__ is defined, str() will fall back to using __repr__. If neither method is defined, Python will use the default representation of the object, which includes the class name and memory address.
class Car():
    def __init__(self,manufacturer, carName):
        self.manufacturer = manufacturer
        self.carName = carName
   # def __str__(self):
    #    return f" {self.carName} is manufactured by {self.manufacturer}"
    def __repr__(self):
        return f" {self.carName} is manufactured by {self.manufacturer}"
```

```python
Car1 = Car("Maruti", "Swift")
str(Car1) # this will call the __str__ method of the Car class and return the string representation of the Car1 object, which is " Swift is manufactured by Maruti".
repr(Car1) # this will call the __repr__ method of the Car class and return the string representation of the Car1 object, which is " Swift is manufactured by Maruti". In this case, since both __str__ and __repr__ methods return the same string, both str() and repr() will give the same output. However, in general, __str__ and __repr__ can return different strings depending on how you want to represent the object for different purposes.
 

''' 

WAHT IS THE USE STR AND REPR METHODS IN PYTHON

__str__() and __repr__() are special methods (dunder methods) in Python used to define how an object is shown as text.

1. __str__() → Human Readable Output

Used when you print an object for users.

Example:

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __str__(self):
        return f"Student Name: {self.name}, Marks: {self.marks}"

s1 = Student("Shubham", 95)

print(s1)

Output:

Student Name: Shubham, Marks: 95

Without __str__():

<__main__.Student object at 0x000001A3>

So __str__() makes objects readable.

2. __repr__() → Developer / Debugging Output

Used to give detailed representation of object for developers.

Example:

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __repr__(self):
        return f"Student('{self.name}', {self.marks})"

s1 = Student("Shubham", 95)

print(repr(s1))

Output:

Student('Shubham', 95)
MAIN DIFFERENCE
Method	Purpose	For
__str__()	Nice readable text	End users
__repr__()	Detailed/debug representation	Developers
IMPORTANT BEHAVIOR

If __str__() is NOT present:

Python automatically uses __repr__().

Example:

class Student:
    def __repr__(self):
        return "This is repr"

s1 = Student()

print(s1)

Output:

This is repr
BEST PRACTICE

Usually:

__str__() → user friendly
__repr__() → debugging friendly

Example together:

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __str__(self):
        return f"Employee Name: {self.name}"

    def __repr__(self):
        return f"Employee('{self.name}', {self.salary})"

e1 = Employee("Shubham", 50000)

print(str(e1))    # uses __str__
print(repr(e1))   # uses __repr__
print(e1)         # uses __str__

Output:

Employee Name: Shubham
Employee('Shubham', 50000)
Employee Name: Shubham
EASY MEMORY TRICK
str → simple string for users
repr → representation for developers
WHERE THEY ARE USED
__str__() used in:
print(obj)
str(obj)
__repr__() used in:
repr(obj)

Also used in debugging, console, lists, etc.

Example:

class Test:
    def __repr__(self):
        return "REPR"

t = Test()

print([t])

Output:

[REPR]

Because lists internally use repr().


'''
```

```python
class Account:
    def __init__(self, balance):
        self.__balance = balance

    # getter
    # def get_balance(self):
    #     return self.__balance

    # setter
    # def set_balance(self, value):
    #     self.__balance = value

    # get, set, del

    # getter
    @property # here we dont swrite the getter because it gets automatically created when we use the @property decorator, we just need to define the method with the same name as the attribute we want to access and it will internally create a getter for that attribute. So we are not directly calling the getter method but it is being called internally when we access the balance attribute.
    def balance(self): #This is an replacement of an getter method,beacause when we want to get the balance we dont nedd to call the gettblanance metthod instead we can directly access the balance attribute and it will internally call the balance method and return the value of the balance. So we are not directly calling the balance method but it is being called internally when we access the balance attribute.
        print("Checking balance")
        return self.__balance
    
    # setter
    @balance.setter # here we need to use setter sice it is the keyword which is used to define a setter method for the balance attribute, it tells python that this method is a setter for the balance attribute and it will be called when we try to set a value to the balance attribute. So we are not directly calling the setter method but it is being called internally when we try to set a value to the balance attribute.
    def balance(self, value):
        if value < 0:
            print("Security Alert. balance cannot be negative")
        elif value > 1000000:
            print("Security Alert. large transaction, requires approval")
        else:
            print("Changing balance")
            self.__balance = value

    @balance.deleter
    def balance(self):
        print("ERROR : You cannot delete a bank balance.")
    
```
