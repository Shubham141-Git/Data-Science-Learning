```python
# str and repr are both special methods in Python that are used to define the string representation of an object, but they serve different purposes. The __str__ method is intended to provide a human-readable representation of the object, while the __repr__ method is intended to provide an unambiguous representation of the object that can be used for debugging and development purposes. If both methods are defined, str() will use __str__, while repr() will use __repr__. If only __repr__ is defined, str() will fall back to using __repr__. If neither method is defined, Python will use the default representation of the object, which includes the class name and memory address.
class Car():
    def __init__(self,manufacturer, carName):
        self.manufacturer = manufacturer
        self.carName = carName
   # def __str__(self):
    #    return f" {self.carName} is manufactured by {self.manufacturer}"
    def __repr__(self):
        return f" {self.carName} is manufactured by {self.manufacturer}"
```

```python
# how to the add the 2 object of the same class, for example we have a class called Wallet and we want to add the cash of two wallet objects, we can define a method called __add__ in the Wallet class which will take another wallet object as an argument and return the sum of the cash of both wallet objects.

class Wallet:
    def __init__(self, cash):
        self.cash = cash
    def __repr__(self):
        return f"Wallet with cash: {self.cash}"
    def __add__(self, other):#here other is an argument which will accept another wallet object and add will add the cash of both wallet objects and return a new wallet object with the total cash.
        total_cash = self.cash + other.cash
        return Wallet(total_cash)
''' 
   Your class is using:

__init__() → constructor
__repr__() → object representation
__add__() → operator overloading (+)

Let's understand line by line.

FULL CODE
class Wallet:
    def __init__(self, cash):
        self.cash = cash
    
    def __repr__(self):
        return f"Class Wallet, Cash : {self.cash}"
    
    def __add__(self, other):
        return Wallet(self.cash + other.cash)
1. Constructor → __init__()
def __init__(self, cash):
    self.cash = cash

Used to initialize object variables.

Example:

w1 = Wallet(50)

Python internally does:

self.cash = 50

So object becomes:

w1.cash = 50
WHAT IS self?

self means:

👉 "current object"

When object calls method:

w1 = Wallet(50)

Python internally sends:

Wallet.__init__(w1, 50)

So:

self = w1
cash = 50
2. __repr__() Method
def __repr__(self):
    return f"Class Wallet, Cash : {self.cash}"

Defines how object should look when printed/debugged.

Example:

w1 = Wallet(50)

print(w1)

Python internally calls:

w1.__repr__()

Returns:

"Class Wallet, Cash : 50"

Output:

Class Wallet, Cash : 50
WITHOUT __repr__()

Output would look ugly:

<__main__.Wallet object at 0x00000123>

So __repr__() makes object readable.

3. __add__() Method
def __add__(self, other):
    return Wallet(self.cash + other.cash)

This enables:

w1 + w2
HOW IT WORKS

Suppose:

w1 = Wallet(50)
w2 = Wallet(30)

Now:

w1 + w2

Python internally converts to:

w1.__add__(w2)

Here:

self = w1
other = w2
INSIDE METHOD
self.cash = 50
other.cash = 30

Then:

50 + 30 = 80

Returns:

Wallet(80)

A NEW object is created.

COMPLETE FLOW
w1 = Wallet(50)
w2 = Wallet(30)

combined = w1 + w2

print(combined)
STEP-BY-STEP
Step 1
w1.cash = 50
Step 2
w2.cash = 30
Step 3
w1 + w2

becomes:

w1.__add__(w2)
Step 4

Returns:

Wallet(80)
Step 5
print(combined)

calls:

combined.__repr__()

Output:

Class Wallet, Cash : 80
IMPORTANT CONCEPT

This is called:

Operator Overloading

Because you are changing behavior of + operator for custom objects.

REAL LIFE IDEA

Imagine:

wallet1 = ₹50
wallet2 = ₹30

Adding both gives:

₹80

Your class mimics this behavior.

MEMORY TRICK
__repr__()
print(object)

↓

object.__repr__()
__add__()
a + b

↓

a.__add__(b)
   
   ''' 
```

```python
def even_numbers(l): #l is a parameter.It receives the list passed to the function.
    final_even_list = []
    for i in l:
        if i%2==0:
            final_even_list.append(i)
    return final_even_list

even_n = even_numbers(numbers)
print(even_n)
```

```python
# Same thing with list comprehension
[i for i in numbers ] # for i in numbers is the loop which iterates through each element in the numbers list and store it in i one by one 
# [i for i in numbers] , “Take every item from numbers and put it into a new list.”

[i for i in numbers if i%2==0] # this will create a new list which will contain only the even numbers from the numbers list. So we are using list comprehension to create a new list which will contain only the even numbers from the numbers list. So we are not directly calling the even_numbers function but it is being called internally when we use the list comprehension to create a new list which will contain only the even numbers from the numbers list.
```

```python
# write a list comprehension from numbers ranging from 1 - 100 and if a number is even
# return true else False

result = [True if i % 2 == 0 else False for i in range(1, 101)]

print(result)

'''   
Ternary Operator (Conditional Expression)
STANDARD FORMAT
value_if_true if condition else value_if_false
'''
```
