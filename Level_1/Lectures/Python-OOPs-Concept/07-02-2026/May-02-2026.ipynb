# Encapsultation

- Encapsulation is the process of wrapping data (variables) and methods (functions) together in a single unit (class) and restricting direct access to the data by using controlled methods.

- Public : syntax--> "name", Access-->anywhere
- Protected : syntax --> "_name", Access --> inside the class and child class
- Private : syntax --> "__name", Access --> only inside class

```python
## Scope Levels in Python

class BankAccount:
    def __init__(self,name,balance):
        self.name=name; #public variable
        self._accountType="savings" #Protected varaible--> Note-->Any Variable in the class that starts with underScore is a protected variable , It is accesssible inside the class,it does not restrict us to use outside the class
        self.__balance=balance; # Private variablr-->Note-->Any variable which starts with double underscore is a private variable , It is accessible inside the class only, not utside the class
       
```

# Inheritance

- Inheritance is the process by which one class (child/subclass) acquires the properties and methods of another class (parent/superclass).

- Allows class to reuse properties and method of another class

- Two types of Inheritace Sinle level inheritance and Multi level inhertance

- Single Inheritance --> Parent --> Child

- Mutilevel Inheritance --> parent --> Child--> child

- Multiple Inheritance --> A + B (parents) --> child



```python
# Single inheritance

class Parent:
    def show_parent(self):
        print("This is Parent class")

class Child(Parent):   # Child inherits from Parent
    def show_child(self):
        print("This is Child class")

# create object
c = Child()

c.show_parent()   # method from Parent
c.show_child()    # method from Child

''' 

How to Explain (Simple Way)

👉 Animal = parent class
👉 Dog = child class

👉 Dog inherits properties of Animal
So:

Dog can eat (from parent)
Dog can bark (its own)

'''

# keywords Pass--> it does not read the line it ignores it
# pass - read and ignore

if True:
     pass
else:
     pass

''' 
The pass keyword in Python is used when you need a statement but don’t want to write any code yet.

🔹 Definition

👉 pass is a null statement — it does nothing.

🔹 Why we use pass

Sometimes Python requires a block of code, but you may not be ready to write it.
Instead of getting an error, you use pass.

In Simple Words

👉 pass means “do nothing for now, I will write code later”
'''
```

```python
# Multilevel Inheritance

class Grandparent:
    def show_grandparent(self):
        print("I am Grandparent")

class Parent(Grandparent):
    def show_parent(self):
        print("I am Parent")

class Child(Parent):
    def show_child(self):
        print("I am Child")

# create object
c = Child()

c.show_grandparent()  # from Grandparent
c.show_parent()       # from Parent
c.show_child()        # from Child

''' How to Explain Simply

👉 Grandparent → Parent → Child (chain)

Parent inherits from Grandparent
Child inherits from Parent
So Child gets all properties of both'''
```

```python
# Multiple Inheritance

class Parent1:
    def show_parent1(self):
        print("This is Parent1")

class Parent2:
    def show_parent2(self):
        print("This is Parent2")

class Child(Parent1, Parent2):   # Child inherits from both
    def show_child(self):
        print("This is Child")

# create object
c = Child()

c.show_parent1()  # from Parent1
c.show_parent2()  # from Parent2
c.show_child()    # own method

''' 
🔹 Simple Explanation

👉 Parent1 + Parent2 → Child

👉 Child gets:

method from Parent1
method from Parent2
its own method

🔥 One-line

👉 Multiple inheritance means one child class inherits from more than one parent class.

'''

#what happens if both parents have same method (important 🔥)
#MRO with simple example

class Parent1:
    def show(self):
        print("Parent1 method")

class Parent2:
    def show(self):
        print("Parent2 method")

class Child(Parent1, Parent2):
    pass

c = Child()
c.show()

'''
🔹 Output
Parent1 method

'''

'''  Why this happens?

👉 Python uses Method Resolution Order (MRO)

👉 It checks classes from left to right

So:

Child(Parent1, Parent2)

➡️ First looks in Parent1
➡️ Then Parent2

That’s why Parent1 method is called

🔹 How to Check MRO
print(Child.mro())
Output:
[Child, Parent1, Parent2, object]
🔹 Simple Rule to Remember

👉 Left side parent gets priority

🔥 Important Tip (Exam)

If order changes:

class Child(Parent2, Parent1):
    pass

👉 Output will be:

Parent2 method
🔹 Final One-Line

👉 MRO decides which method is called when multiple parents have same method.'''

#Analogy in inheritance -->Used for Design concept
#IS A relationship
#HAS A relationship
```

# Method Overriding
 - same name
 - when a child class has a method with the same name as the method in the parent class, then the child class method replaces (overrides) the parent class method.
 

```python
# parent class
class Parent:
    def show(self):
        print("This is parent class method")


# child class
class Child(Parent):
    def show(self): # here we are overriding the show method of parent class in child class
        super().show() # calling parent class method using super() --->Internally Super means object of paarent class --> internaly super() creates object of parent class and calls the method of parent class
        #Parent().show() --> another way to call parent class method using class name and creating object of parent class
        print("This is child class method")


# object of child class
c = Child()

# calling method
c.show()


'''Here we have a parent and child class with same method name show() when we create the object of child class and call the show method it will call the child class method because of method overriding in python

Override itseld means that theat old metgod
 will be overridden by new method with same 
 name and same signature in the child class 
 when we call the method using child class
   object it will call the child class
     method instead of parent class method
       this concept is called method 
       overriding in python
'''


''' explanation step by step
1. parent class created
class Parent:

inside it:

def show(self):

this method prints:

This is parent class method
2. child class inherits parent
class Child(Parent):

this means:

child class gets all properties and methods of parent class
3. child creates same method again
def show(self):

same method name as parent.

this is called method overriding.

the child method replaces parent method.

4. object created
c = Child()

object belongs to child class.

5. method called
c.show()

python first checks inside Child class.

it finds:

def show(self):

so it runs child method, not parent method.

therefore output is:

This is child class method
memory flow
c.show()

↓

python searches in Child

↓

method found in Child

↓

parent method ignored

↓

child method executes

using parent method also with super()
class Parent:
    def show(self):
        print("Parent method")


class Child(Parent):
    def show(self):
        super().show()
        print("Child method")


c = Child()
c.show()
output
Parent method
Child method
how super() works
super().show()

means:

call the show() method of parent class.'''


#Super() keyword isused when child and parent has the same method and when childclass overirides themethod in parent class but user want to call the method present in parent class then we can use super() keyword to call the method of parent class from child class
```

# MRO -->Method resolution order
- used to find next method in inheritance
-it tells python:in which order python searches methods and variables in inheritance.

- especially important in:
  multiple inheritance
   multilevel inheritance

```python
class A:
    def __init__(self):
        print("Class A")
    def a(self):
        print("a")

class B(A):
    def __init__(self):
        super().__init__()
        print("Class B")
    def b(self):
        print("b")

class C(A):
    def __init__(self):
        print("Class C")
    def c(self):
        print("c")

class D(B,C): # not available in java # ambiguity
    def __init__(self):
        super().__init__()
        print("Class D")
    def d(self):
        print("d")
```

```python
d_obj = D()
print(D.mro())# Used to check the method resolution order in case of multiple inheritance

```

```python
from abc import ABC, abstractmethod #importing ABC and abstractmethod from abc module

class Payment(ABC): # making Payment an abstract class by inheriting ABC i.e ABC will make the class abstract class

    @abstractmethod # to make the methods abstract method we use @abstractmethod decorator
    def pay(self, amount):
        pass # abstract method with no implementation
    def refund(self, amount):
        pass # another abstract method with no implementation

# Here Methods pay() and refund() are abstract methods because they are decorated with @abstractmethod and they have no implementation in the parent class Payment
#Class who will inherit the Payment class will be forced to implement the pay() and refund() method otherwise it will give error because of abstract method
#pay n refund are the  incomplete methods and they are just a blueprint for the child class to implement them according to their need
```

```python
class IDBIBANK(Payment): # IDBIBANK is inheriting the Payment class so it will be forced to implement the pay() and refund() method because they are abstract methods in the parent class

 def newUser(self):
        print("Welcome to IDBI Bank")
    #But here we are not implementing the pay() and refund() method in the child class IDBIBANK so it will give error because of abstract method in parent class Payment    
    #
```
