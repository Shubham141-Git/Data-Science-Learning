## Function arguments

```python
#Default argument

'''A default argument is a parameter that already has a predefined value in a function.

If the user does not provide a value for that parameter while calling the function, Python automatically uses the default value.

This helps prevent errors when some arguments are missing and makes the function more flexible.
'''

def sum(a=0,b=0): 
    return a+b
print(sum(1,3))   

def sum(a,b=0): # Note--> Rule==> parameter without the default should be decaled first and then after paramater with default value should be created
    return a+b
print(sum(1,3))    
```

```python
#Using programming languages we can read file, update file, append more content to file, close file

#To open the file we make use of open() which che k from current directory

f=open("dummy.txt")
type(f)

#To read all content of file at once we use read()--> read everyhing as 1 big string

newline=f.read().split("\n") # "\n" stand for new line
print(newline)


#To close file we use close()
f.close()

#to read line --> readlines -->read as everything as a list of string

g=open("dummy.txt")
lines=g.readlines()
print(lines)
g.close()

#to read line by line not all at once--> readline--> read just 1 line at a time

g=open("dummy.txt")
line=g.readline()
print(line)



line=g.readline()
print(line)



line=g.readline()
print(line)
g.close()

'''

🔹 Difference Between readline() and readlines()
Feature	readline()	readlines()
What it reads	One line at a time	All lines at once
Return type	String	List of strings
Memory usage	Low (good for large files)	Higher (loads full file)
Use case	Looping line by line	When you need all lines together
✅ 1️⃣ readline()
🔹 Reads only ONE line from file

Example file (data.txt):

Apple
Banana
Mango

Example code:
f = open("data.txt", "r")

line1 = f.readline()
print(line1)

line2 = f.readline()
print(line2)

f.close()

Output:
Apple
Banana


👉 Every time you call readline(), it reads the next line.

✅ 2️⃣ readlines()
🔹 Reads ALL lines at once

Example:

f = open("data.txt", "r")

lines = f.readlines()
print(lines)

f.close()

Output:
['Apple\n', 'Banana\n', 'Mango\n']


👉 It returns a list
👉 Each line is a string
👉 \n means newline

🔥 Important Difference in Return Type
type(f.readline())    # str
type(f.readlines())   # list

🧠 Best Practice (Very Important)

Instead of using both, the best way is:

with open("data.txt", "r") as f:
    for line in f:
        print(line.strip())


✔ Cleaner
✔ Memory efficient
✔ Automatically closes file

🚀 When To Use What?
Use readline() when:

File is very large

You want to process line by line

Use readlines() when:

File is small

You need all lines together

'''


#note--> if we hace 8gb of ram and the file size is 12gb
##if we use read/readlines function when systems will not be able to read the file due to memory contraints
#so at such time we can make use of readline function which will read one line at a time and there will be no load on system

```

```python
#to read line one by one we can also make use of for loop
h=open("dummy.txt")
while True:
    line=h.readline()
    if not line: #--> this means if there is no line to read then break the loop
        break
    print(line.strip()) # strip() removes the newline character at the end of each line
h.close()


#Note--> it is possiblr that we may not close file due to some error or we may forget to close the file, so to avoid such situation we can make use of "with" keyword which will automatically close the file once the block of code is executed it is called as "context manager"  which will automatically close the file once the block of code is executed
with open("dummy.txt") as f: #here dummy.txt is relative path since it is in current working directory
    for line in f:
        print(line.strip())

# To get the full path of file we can make use of os.path.abspath() which will give us the full path of file
import os
y=os.path.abspath("dummy.txt")  
print(y)      
```
