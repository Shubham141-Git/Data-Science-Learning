## Strings Method


```python
#Upper()
str="  Data science is fun  "
str1=str.upper()
print(str)
print(str1)

##lower
str2=str.lower()
print(str2)

#Title--> first character of evry word is capital

str3=str.title()
print(str3)

#Capitalize --> only capitalize first word
str4=str.capitalize()
print(str4)

#Swapcase --> make smalln to capital n capital to small

str5=str.swapcase()
print(str5)


#strip--> removes spcess
str6=str.strip()
print(str6)

#split--> will return the element in an list/array
newstring="123,err,the,tsfs,vfdfd"
ss=newstring.split(",")
print(ss)


#Find --> will return the index of the ccharactter
str7=str.find("s")
print(str7)


#join --> combine the string into one/also list elements

str8=str.split(" ")

print(" ".join(str8)) #--> " " will add space after every word like a seperator


#Sorted() ---> sort the 

a=["z","q",a","b","c"]

b=sorted(a) #---> it makes asc

c=sorted(a, reverse=True) #--> it makes descending
```

## While Loop

```python
# Keep adding 1 until count is <=5

count=0

while count<=5:
   print(count)
   count+=1
   print(count)

```

## Control Flow

```python
## Continue --> skips the partuclar iteration

##USe Case to skip the iteratio if the nale is monal

name=["Shubham","Akash","Monal","Suraj"]

for a in name:
    if a=="Monal":
        continue
    a=a.upper()
    print(a)

 

print("------")

#Break-->    terminates  the loop itself 

##USe terminate the iteratio if the nale is monal

name=["Shubham","Akash","Monal","Suraj"]

for a in name:
    if a=="Monal":
        break
    a=a.upper()
    print(a)

#UseCase print 1 to 100 and break the loop when valuw>100

count=0

while True:
    if count>100:
    break

  
```
